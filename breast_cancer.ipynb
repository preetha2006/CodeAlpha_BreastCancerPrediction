Load the necessary packages

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

Load the dataset

```python
df = pd.read_csv('/content/drive/MyDrive/Datasets/codealpha/breast cancer dataset.csv')
df
```

Drop "id" column

```python
df = df.drop(['id'], axis=1)

# Features and target
X = df.drop(['diagnosis'], axis=1)
y = df['diagnosis']
```

```python
# Encode target ('M'=1, 'B'=0)
le = LabelEncoder()
y_enc = le.fit_transform(y)
```

```python
# Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_enc, test_size=0.2, random_state=42, stratify=y_enc)
```

```python
# Model training (Random Forest example)
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)
```

```python
# Prediction and metrics
y_pred = clf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))
```

```python
import gradio as gr
import numpy as np


# Dummy breast cancer prediction function (replace with your model inference)
def breast_cancer_predict(radius_mean, texture_mean, perimeter_mean, area_mean, smoothness_mean):
    # Simple risk calculation (demo purposes)
    risk_score = (
        0.3 * radius_mean / 30 +
        0.25 * texture_mean / 40 +
        0.2 * perimeter_mean / 200 +
        0.15 * area_mean / 2500 +
        0.1 * smoothness_mean / 0.2
    )
    prob = min(risk_score, 1.0)
    if prob > 0.7:
        risk = "High Risk"
    elif prob > 0.4:
        risk = "Moderate Risk"
    else:
        risk = "Low Risk"
    return f"Breast Cancer Risk Probability: {prob:.2f} ({risk})", prob


css = """
.gradio-container {
    max-width: 700px;
    margin: auto;
    background:#fbeed8;
    border-radius:12px;
    padding:20px;
    font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
}
.gr-button {
    background:#af9972;
    color:#fff;
    font-weight:700;
}
.gr-button:hover {
    background:#9b865e;
}
#risk_meter {
    margin-top: 1rem;
}
"""


with gr.Blocks(css=css) as demo:
    gr.Markdown(
        "<h2 style='text-align:center; font-weight:bold; font-family: Segoe UI, Tahoma, Geneva, Verdana, sans-serif;'>Breast Cancer Prediction</h2>"
    )

    with gr.Row():
        radius_mean = gr.Number(
            label="Radius Mean",
            precision=2,
            interactive=True,
            elem_id="radius_mean",
            info="Typically range: 6 to 28. Provide approximate tumor radius mean.",
        )
        texture_mean = gr.Number(
            label="Texture Mean",
            precision=2,
            interactive=True,
            info="Example values: 9 to 40. Represents variation in gray-scale intensity.",
        )

    with gr.Row():
        perimeter_mean = gr.Number(
            label="Perimeter Mean",
            precision=2,
            interactive=True,
            info="Typical range: 40 to 190. Length around tumor boundary.",
        )
        area_mean = gr.Number(
            label="Area Mean",
            precision=2,
            interactive=True,
            info="Approximate range: 150 to 2500. Size measurement of tumor area.",
        )

    smoothness_mean = gr.Number(
        label="Smoothness Mean",
        precision=4,
        interactive=True,
        info="Range: roughly 0.05 to 0.2. Measures tumor edge smoothness.",
    )

    output_text = gr.Textbox(label="Prediction Output", interactive=False)
    risk_meter = gr.Number(label="Risk Probability Meter", interactive=False, elem_id="risk_meter")

    def validate_and_predict(radius, texture, perimeter, area, smoothness):
        # Input validation example
        if radius is None or texture is None or perimeter is None or area is None or smoothness is None:
            return "Please enter values for all fields.", 0.0
        if radius < 0 or texture < 0 or perimeter < 0 or area < 0 or smoothness < 0:
            return "All inputs must be non-negative numbers.", 0.0

        return breast_cancer_predict(radius, texture, perimeter, area, smoothness)

    predict_btn = gr.Button("Predict")
    predict_btn.click(
        fn=validate_and_predict,
        inputs=[radius_mean, texture_mean, perimeter_mean, area_mean, smoothness_mean],
        outputs=[output_text, risk_meter],
    )


demo.launch()
```
